```python
import matplotlib.pyplot as plt
from astropy.visualization import make_lupton_rgb
from astropy.io import fits
from astropy.utils.data import get_pkg_data_filename
```

```python

# Read in the three images downloaded from here:
r_f160w = get_pkg_data_filename('hlsp_hudf12_hst_wfc3ir_udfmain_f160w_v1.0_drz.fits')
g_f125w = get_pkg_data_filename('hlsp_hudf12_hst_wfc3ir_udfmain_f125w_v1.0_drz.fits')
b_f105w = get_pkg_data_filename('hlsp_hudf12_hst_wfc3ir_udfmain_f105w_v1.0_drz.fits')
r = fits.open(r_f160w)[0].data
g = fits.open(g_f125w)[0].data
b = fits.open(b_f105w)[0].data

rgb_default = make_lupton_rgb(r,g,b,Q=20,stretch=0.1,filename="final_colorized.jpeg")
plt.imshow(rgb_default, origin='lower')
plt.savefig("RGB_rojenkin.png")
```
